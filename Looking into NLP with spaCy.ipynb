```python
import spacy
from spacy import displacy
```

```python
nlp = spacy.load("en_core_web_sm")
```

```python
doc = nlp("He must contend with National Assembly elections in the next two months and the prospects of spring labor unrest and renewed campus protests by college students, now on a long winter's recess.")
```

```python
for token in doc:
    print(token.text, token.lemma_, token.pos_, token.tag_, token.dep_,
            token.shape_, token.is_alpha, token.is_stop)
```

```python
displacy.render(doc, style="dep")
```
