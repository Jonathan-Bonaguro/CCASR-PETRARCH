```python
import spacy
from spacy import displacy
```

```python
spacy.prefer_gpu()
nlp = spacy.load("en")
```

```python
doc = nlp('The mood during the voting, under cold gray skies Sunday, was quiet, despite reports of clashes between riot policemen and demonstrators in several cities.')
```

```python
for token in doc:
    print(token.text, token.lemma_, token.pos_, token.tag_, token.dep_,
            token.shape_, token.is_alpha, token.is_stop)
```

```python
displacy.render(doc, style="dep")
```
