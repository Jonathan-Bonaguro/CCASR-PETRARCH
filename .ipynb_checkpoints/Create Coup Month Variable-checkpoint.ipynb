```python
df = pd.read_csv("bs_CDP_rep.csv", usecols=['cowcode', 'year', 'CDP'])
```

```python
df['CDP'].value_counts()
```

```python
df
```

```python
def custcum(x):
    total = 1
    for i, v in x.iterrows():
        x.loc[i, 'yrs_since_CDP'] = total
        total += 1
        if v.CDP in [1, 2]:
            total = 0
    return x
```

```python
df_gb = df.groupby(['cowcode']).apply(custcum)
```

```python
df_gb.head(500)
```

```python
df_gb.to_csv("with_CDP_years_since_var.csv", index=False)
```
