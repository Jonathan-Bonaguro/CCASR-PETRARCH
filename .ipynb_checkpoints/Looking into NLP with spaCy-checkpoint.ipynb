```python
import spacy
from spacy import displacy
```

```python
nlp = spacy.load("en_core_web_sm")
```

```python
doc = nlp("North Korea has only admitted to abducting 13 Japanese, but a citizens' group made public this week a list of 143 possible abductees.")
```

```python
for token in doc:
    print(token.text, token.lemma_, token.pos_, token.tag_, token.dep_,
            token.shape_, token.is_alpha, token.is_stop)
```

```python
displacy.render(doc, style="dep")
```
